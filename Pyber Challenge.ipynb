### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import os


# File to Load (Remember to change these)
city_data = os.path.join('.', 'Resources', 'city_data.csv')
ride_data = os.path.join('.', 'Resources', 'ride_data.csv')

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data)
ride_data_df = pd.read_csv(ride_data)
```

```python
city_data_df['city'].nunique()
```

### Merge the DataFrames

```python
# Combine the data into a single dataset (2375)
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city"])

# Display the data table for preview
pyber_data_df.head(20)
```

```python
pyber_data_df.describe()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

totalrides = pyber_data_df.groupby(['type']).count()['ride_id'].rename("Total Rides")
totalrides
```

```python
# 2. Get the total drivers for each city type

driver_count = pyber_data_df.groupby(['city','type']).max(['driver_count'])
totaldrivers = driver_count.groupby('type').sum()['driver_count'].rename("Total Drivers")
totaldrivers
```

```python
#  3. Get the total amount of fares for each city type

totalfares = pyber_data_df.groupby(['type']).sum()['fare'].rename("Total Fares")
totalfares
```

```python
#  4. Get the average fare per ride for each city type. 

avgfare_per_ride = pyber_data_df.groupby(['type']).mean()['fare'].rename("Avg Fare per Ride")
avgfare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 

avg_fare_df = pd.merge(totaldrivers, totalfares, how='left', on='type')
avgfare_per_driver = (avg_fare_df['Total Fares']/avg_fare_df['Total Drivers']).rename("Avg Fare per Driver")
avgfare_per_driver
```

```python
avg_fare_df.head()
```

```python
#  6. Create a PyBer summary DataFrame. 

df_totalrides = pd.DataFrame(totalrides)
df_totaldrivers= pd.DataFrame(totaldrivers)
df_totalfares= pd.DataFrame(totalfares)
df_avgfare_per_ride= pd.DataFrame(avgfare_per_ride)
df_avgfare_per_driver= pd.DataFrame(avgfare_per_driver)

pyber_summary_df = pd.concat([df_totalrides, df_totaldrivers, df_totalfares, df_avgfare_per_ride, df_avgfare_per_driver], axis=1)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name

pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.

fmt_df = pyber_summary_df.copy()

fmt_df.loc[:, "Total Rides"] = fmt_df["Total Rides"].map("{:,d}".format)
fmt_df.loc[:, "Total Drivers"] = fmt_df["Total Drivers"].map("{:,d}".format)
fmt_df.loc[:, "Total Fares"] = fmt_df["Total Fares"].map("${:,.2f}".format)
fmt_df.loc[:, "Avg Fare per Ride"] = fmt_df["Avg Fare per Ride"].map("${:,.2f}".format)
fmt_df.loc[:, "Avg Fare per Driver"] = fmt_df["Avg Fare per Driver"].map("${:,.2f}".format)

fmt_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame

pyber_summary_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

fare_by_datetype_df = pyber_data_df.groupby(['type', 'date']).sum()[['fare']]
fare_by_datetype_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fare_by_datetype_df = fare_by_datetype_df.reset_index()
fare_by_datetype_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date.

fare_by_datetype_pivot = fare_by_datetype_df.pivot(index='date', columns='type', values='fare')
fare_by_datetype_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

date_range_df=fare_by_datetype_pivot.loc['2019-01-01':'2019-04-29']
date_range_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
date_range_df.index = pd.to_datetime(date_range_df.index)
date_range_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
date_range_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.

fare_by_week = date_range_df.resample('W').sum()
fare_by_week
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

legend = ['Rural','Suburban','Urban']

fig, ax = plt.subplots(figsize=(12,5))
ax.plot(fare_by_week, label = legend)
ax.set_ylim(0,2600)
#ax.set_xlim(0,100)
ax.set_ylabel('Fare ($USD)', fontsize=12)
ax.set_title("Total Fares by City Type", fontsize=15)
ax.legend(title = 'type', loc = 'center', fontsize=12)

import matplotlib.dates as mdates
date_format = mdates.DateFormatter('%b')
plt.gca().xaxis.set_major_formatter(date_format) 
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))
plt.show()
plt.savefig("analysis/Pyber_fare_summary.png")
```
